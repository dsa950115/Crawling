```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

url = 'https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2'

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36 Edg/85.0.564.44'}

res = req.get(url, headers=headers)

soup = bs(res.content, 'lxml')

soup
```

```python
rank = []
title = []
artist = []

for i in soup.select("td.ranking"):
    rank.append(i.text)
    
for i in soup.select("a.title > span"):
    title.append(i.text)
    
for i in soup.select("a._artist > span"):
    artist.append(i.text.split())
    
music = pd.DataFrame({"rank" : rank, "title" : title, "artist" : artist})
```

```python
music = music.set_index('rank')
music
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200907"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36"}

res = req.get(url, headers = headers)
soup = bs(res.content, 'lxml')
soup
```

```python
title = []
score = []

for i in soup.select("div.tit5 > a"):
    title.append(i.text)
    
for i in soup.select("td.point"):
    score.append(i.text)
    
num = [i for i in range(1,len(title)+1)]

pd.DataFrame({"num" : num, "title" : title, "score" : score})
```
