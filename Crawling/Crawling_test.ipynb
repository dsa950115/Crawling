# 문제 2

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}

res = req.get(url, headers = headers)
soup = bs(res.content, "lxml")
soup
```

```python
movie_list = []
m = soup.select("td.title div a")

for i in m:
    movie_list.append(i.text)

rank_list = list(range(1,len(movie_list)+1))

df = pd.DataFrame({"rank" : rank_list, "movie" : movie_list})
df.set_index("rank", inplace=True)
df
```

# 문제 3

```python
url = "https://front.wemakeprice.com/best"

driver = wd.Chrome()
driver.get(url)
```

```python
name_list = []
price_list = []

for index in range(0,10,1):
    #image_list = driver.find_elements_by_css_selector('div.flag_wrap')
    lst = driver.find_elements_by_css_selector('p.text')
    time.sleep(0.5)
    #image_list[index].click()
    lst[index].click()
    time.sleep(0.5)
    
    #크롤링(제품 타이틀, 가격, 분류, 배송)
    
    try:
        name = driver.find_element_by_css_selector('h3.deal_tit').text
    except:
        name = -1
        print("제품 이름 없음")
        
    try:
        price = driver.find_element_by_css_selector('strong.sale_price em.num').text
    except:
        price = -1
        print("제품 가격 없음")
        
     
    name_list.append(name)
    price_list.append(price)

    driver.back()
    time.sleep(2)

rank_list = [i for i in range(1,11)]
    
df = pd.DataFrame({"rank" : rank_list ,"name" : name_list, "price" : price_list})
df.set_index("rank", inplace = True)
```

```python
df.to_csv("wemakeprice.csv", header = True, encoding = "EUC-KR")
```

# 문제 4

```python
url = "https://www.koreabaseball.com/teamrank/teamrank.aspx"

res = req.get(url)
soup = bs(res.content, 'lxml')
soup
```

```python
ans = []

for i in soup.select("table.tData tr"):
    a = i.text.split("\n")[1:-1]
    if a:
        ans.append(a)

ans = np.array(ans)

df = pd.DataFrame(ans[1:,:], columns = ans[0,:])
df = df.iloc[:, :7]

df.set_index("순위", inplace = True)
df
```

```python
df.to_csv("baseball.csv", header = True, encoding = "EUC-KR")
```
