```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook
```

# 11번가 크롤링

```python
url = "http://www.11st.co.kr/browsing/BestSeller.tmall?method=getBestSellerMain"

driver = wb.Chrome()
driver.get(url)
```

```python
name_list = []
price_list = []
cate_list = []
ship_list = []


for index in range(1,11,1):
    image_list = driver.find_elements_by_css_selector('span.best + div.img_plot')
    time.sleep(1)
    image_list[index].click()
    time.sleep(1)
    
    #크롤링(제품 타이틀, 가격, 분류, 배송)
    
    
    
    try:
        name = driver.find_element_by_css_selector('h1.title').text
    except:
        name = np.NaN
        print("제품 이름 없음")
        
    try:
        price = driver.find_element_by_css_selector('strong > span.value').text
    except:
        price = np.NaN
        print("제품 가격 없음")
        
    try:
        cate = driver.find_element_by_css_selector('em.selected').text
    except:
        cate = np.NaN
        print("제품 분류 없음")
    
    try:
        ship = driver.find_element_by_css_selector('div.delivery > dt').text
    except:
        ship = np.NaN
        print("무료배송 아님")
     
    name_list.append(name)
    price_list.append(price)
    cate_list.append(cate)
    ship_list.append(ship)
       
    driver.back()
    time.sleep(2)
    
pd.DataFrame({"name" : name_list, "price" : price_list, "cate" : cate_list, "ship" : ship_list})
```
